```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Introduccin al truncamiento de distribuciones
"""
El truncamiento de distribuciones es una tcnica utilizada en estadstica cuando queremos
limitar el rango de valores que una variable aleatoria puede tomar. Es til cuando sabemos
que los datos reales solo pueden existir dentro de ciertos lmites.

Por ejemplo, si estamos midiendo alturas de personas, no tendra sentido permitir valores negativos.

En este notebook, exploraremos cmo truncar una distribucin y ajustarla a datos reales.
"""

# Definimos una distribucin normal sin truncamiento
mu, sigma = 0, 1  # Media y desviacin estndar
x = np.linspace(-4, 4, 1000)
pdf_original = stats.norm.pdf(x, mu, sigma)

# Definimos los lmites de truncamiento
lower, upper = -1.5, 1.5  # Rango permitido
trunc_norm = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma
)

# Creamos valores dentro del rango truncado
t_x = np.linspace(lower, upper, 1000)
t_pdf = trunc_norm.pdf(t_x)

# Graficamos la distribucin original y la truncada
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_original, label='Distribucin Normal Original', linestyle='dashed')
plt.plot(t_x, t_pdf, label='Distribucin Truncada', color='red')
plt.axvline(lower, color='black', linestyle='dotted', label='Lmite Inferior')
plt.axvline(upper, color='black', linestyle='dotted', label='Lmite Superior')
plt.legend()
plt.title("Ejemplo de Truncamiento de una Distribucin Normal")
plt.xlabel("Valores de la variable")
plt.ylabel("Densidad de probabilidad")
plt.show()

# Espacio para agregar un ejemplo prctico personalizado
```
