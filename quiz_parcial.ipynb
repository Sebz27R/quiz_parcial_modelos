```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Introducción al truncamiento de distribuciones
"""
El truncamiento de distribuciones es una técnica utilizada en estadística cuando queremos
limitar el rango de valores que una variable aleatoria puede tomar. Es útil cuando sabemos
que los datos reales solo pueden existir dentro de ciertos límites.

Por ejemplo, si estamos midiendo alturas de personas, no tendría sentido permitir valores negativos.

En este notebook, exploraremos cómo truncar una distribución y ajustarla a datos reales.
"""

# Definimos una distribución normal sin truncamiento
mu, sigma = 0, 1  # Media y desviación estándar
x = np.linspace(-4, 4, 1000)
pdf_original = stats.norm.pdf(x, mu, sigma)

# Definimos los límites de truncamiento
lower, upper = -1.5, 1.5  # Rango permitido
trunc_norm = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma
)

# Creamos valores dentro del rango truncado
t_x = np.linspace(lower, upper, 1000)
t_pdf = trunc_norm.pdf(t_x)

# Graficamos la distribución original y la truncada
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_original, label='Distribución Normal Original', linestyle='dashed')
plt.plot(t_x, t_pdf, label='Distribución Truncada', color='red')
plt.axvline(lower, color='black', linestyle='dotted', label='Límite Inferior')
plt.axvline(upper, color='black', linestyle='dotted', label='Límite Superior')
plt.legend()
plt.title("Ejemplo de Truncamiento de una Distribución Normal")
plt.xlabel("Valores de la variable")
plt.ylabel("Densidad de probabilidad")
plt.show()

# Espacio para agregar un ejemplo práctico personalizado
```
